### namedtuples
a convenience structure simpler than classes but more readable than tuples or dictionaries,

#### Usage
Person = namedtuple('Person', 'first middle last')

p = Person('Timothy', 'Roger', 'Slator')
p.first
p.middle
p.last


### deque
double-ended queue supports append and pop operations from both ends
* can simulate real workld environments like lines
* can be used to process items with higher priority at one end than the other 

#### Methods
appendleft, popleft, reverse, rotate

usage: dq = deque(iterable, max_entries)

### defaultDict
collection that instead of raising a KeyError when an invalid key is requested, it invokes a provided function and returns a value from that function

d = defaultDict(str)
d['greet'] = 'hello'
print(d['greet'])
print(d['greet1'])

### Counter
items = [1, 2, 3, 4, 5, 6]
most_common = Counter(items).most_common(1)
